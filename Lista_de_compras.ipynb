# **Lista de Compras**


```python
lista_compras = []
```

```python
def adicionar_item():
    nome = input("Digite o nome do item: ").strip()
    if len(nome) == 0:
        print("Você não digitou nenhum nome ")
        return

    qtd_txt = input("Quantidade (inteiro > 0): ").strip()
    try:
        qtd = int(qtd_txt)
    except ValueError:
        print("Quantidade deve ser um número inteiro.")
        return

    if qtd <= 0:
      print("Quantidade deve ser maior que 0")
      return

    item = {"nome": nome, "quantidade": qtd, "comprado": False}
    lista_compras.append(item)
    print("Item adicionado!")
```

```python
def listar_itens():
    if len(lista_compras) == 0:
        print("Lista vazia.")

    print("\nItens da lista:")
    for i, item in enumerate(lista_compras, start=1):
        status = "Comprado" if item["comprado"] else "Não comprado"
        print(f"{i}. {item['nome']}  (qtd: {item['quantidade']})  - {status}")
```

```python
def marcar_comprado():
    if len(lista_compras) == 0:
        print("Lista vazia.")
        return

    # Mostra a lista com índices para escolher
    for i, item in enumerate(lista_compras, start=1):
        status = "Comprado" if item["comprado"] else "Não comprado"
        print(f"{i}. {item['nome']} (qtd: {item['quantidade']}) - {status}")

    idx_txt = input("Digite o número do item a marcar como comprado: ").strip()
    try:
        idx = int(idx_txt)
    except ValueError:
        print("Digite um número inteiro válido.")
        return

    if idx < 1 or idx > len(lista_compras):
        print("Índice fora do intervalo.")
        return

    lista_compras[idx - 1]["comprado"] = True
    print("Item marcado como comprado!")
```

```python
def remover_item():
    if len(lista_compras) == 0:
        print("Lista vazia.")
        return

    # Mostra a lista para escolher
    for i, item in enumerate(lista_compras, start=1):
        status = "Comprado" if item["comprado"] else "Não comprado"
        print(f"{i}. {item['nome']} (qtd: {item['quantidade']}) - {status}")

    idx_txt = input("Digite o número do item para remover: ").strip()
    try:
        idx = int(idx_txt)
    except ValueError:
        print("Digite um número inteiro válido.")
        return

    if idx < 1 or idx > len(lista_compras):
        print("Índice fora do intervalo.")
        return

    removido = lista_compras.pop(idx - 1)
    print(f"Item removido: {removido['nome']}")
```

# Desafios extras (opcional)

```python
def editar_quantidade():
    if len(lista_compras) == 0:
        print("Lista vazia.")
        return

    for i, item in enumerate(lista_compras, start=1):
        status = "Comprado" if item["comprado"] else "Não comprado"
        print(f"{i}. {item['nome']} (qtd: {item['quantidade']}) - {status}")

    idx_txt = input("Número do item para editar a quantidade: ").strip()
    try:
        idx = int(idx_txt)
    except ValueError:
        print("Digite um número inteiro válido.")
        return

    if idx < 1 or idx > len(lista_compras):
        print("Índice fora do intervalo.")
        return

    qtd_txt = input("Nova quantidade (inteiro > 0): ").strip()
    try:
        qtd = int(qtd_txt)
    except ValueError:
        print("Quantidade deve ser um número inteiro.")
        return

    if qtd <= 0:
        print("Quantidade deve ser maior que zero.")
        return

    lista_compras[idx - 1]["quantidade"] = qtd
    print("Quantidade atualizada!")
```

```python
def listar_nao_comprados():
    nao = [it for it in lista_compras if not it["comprado"]]
    if len(nao) == 0:
        print("Não há itens pendentes.")
        return
    print("\nItens não comprados:")
    for i, item in enumerate(nao, start=1):
        print(f"{i}. {item['nome']}  (qtd: {item['quantidade']})")
```

```python
def main():
    while True:
        print("\nMenu:")
        print("1. Adicionar item")
        print("2. Listar itens")
        print("3. Marcar item como comprado")
        print("4. Remover item")
        print("5. Sair")
        print("6. Editar quantidade")          # extra
        print("7. Listar apenas não comprados") # extra
        opc = input("Escolha: ")

        if opc == "1":
            adicionar_item()
        elif opc == "2":
            listar_itens()
        elif opc == "3":
            marcar_comprado()
        elif opc == "4":
            remover_item()
        elif opc == "5":
            print("Você saiu.")
            break
        elif opc == "6":
            editar_quantidade()
        elif opc == "7":
            listar_nao_comprados()
        else:
            print("Opção inválida")
```

```python
main()
```
